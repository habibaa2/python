```python
df = pd.DataFrame({'Publisher': np.random.choice([*'Publisher'], 5000),
                   'Global_Sales': np.random.randint(0, 2, 5000)})
crosstab_df = pd.crosstab(df['Publisher'], df['Global_Sales'], normalize='index').mul(100).sort_values(by=1, ascending=False)
ax = crosstab_df[1].plot.bar()
ax.bar_label(ax.containers[0], fmt='%.2f %%')
ax.margins(y=0.1)
plt.show()
```
